```python
import networkx as nx
import random
```

```python
G1 = nx.read_edgelist("out.txt", nodetype=int, create_using=nx.DiGraph, edgetype=float)
G2 = nx.read_edgelist('musae.txt', create_using=nx.Graph, nodetype=int, edgetype=float).to_directed()
```

```python
for node in G1.nodes:
    sums = 0
    for n in G1.neighbors(node):
        sums += G1[node][n]['weight']
    print(node)
    print(sums)
```

```python
alpha = 0.15

def comparePageRank(G1, G2, n, k):
    G1_personalization = dict()
    for node in G1.nodes:
        G1_personalization[node] = 0
    G1_personalization[n-1] = 1

    G1_pagerank = nx.pagerank(G1, 1-alpha, G1_personalization, max_iter=500, tol=1e-16)

    G1_partial_sum = sum(v for kk, v in G1_pagerank.items() if kk >= n-k)
    for kk, v in G1_pagerank.items():
        G1_pagerank[kk] /= G1_partial_sum

    G2_personalization = dict()
    for node in G2.nodes:
        G2_personalization[node] = 0
    G2_personalization[k-1] = 1

    G2_pagerank = nx.pagerank(G2, 1-alpha, G2_personalization, max_iter=500, tol=1e-16)

    G2_partial_sum = sum(v for kk, v in G2_pagerank.items() if kk < k)
    for kk, v in G2_pagerank.items():
        G2_pagerank[kk] /= G2_partial_sum

    for i in range(k):
         print(i, G1_pagerank[n-k+i], G2_pagerank[i])
```

```python
comparePageRank(G2, G1, 37600, 100)
```
